## **Overview**

This project shall analyze data from the National 2009 H1N1 Flu Survey to understand the factors influencing vaccination patterns. The primary goal is to predict whether individuals received the H1N1 flu vaccine based on various demographic, opinion-based, and health-related features collected in the survey.

## **Business Understanding**

Vaccination is a critical component of public health strategy to control infectious diseases like H1N1 and seasonal flu. Despite the availability of vaccines, uptake rates can vary significantly across different population segments. Understanding the determinants of vaccine acceptance can help public health officials design more effective vaccination campaigns, target high-risk groups, and ultimately improve vaccination rates.


Key stakeholders include:

Public Health Officials: Use the insights to design targeted interventions and communication strategies.
Healthcare Providers: Identify and address barriers to vaccination among patients.
Policy Makers: Develop policies to increase vaccine coverage and accessibility.
Researchers: Study the impact of various factors on vaccine uptake to inform future research.

## **Problem Statement** 

In the context of ongoing efforts to understand and improve vaccination rates, this project aims to predict whether individuals received the H1N1 flu vaccine using data collected from the National 2009 H1N1 Flu Survey. The goal is to develop a binary classification model that accurately identifies individuals who received the H1N1 vaccine based on their demographic information, health behaviors, opinions, and other relevant factors.

By identifying key predictors of H1N1 vaccination, the project seeks to provide actionable insights that can inform public health strategies, optimize resource allocation, and enhance future vaccination campaigns to achieve higher immunization rates and better public health outcomes.

1. **Drop columns** 

```python
# Lets first drop some columns which do not add value to our analysis and may introduce noise. 
columns_to_drop = ['respondent_id', 'employment_industry', 'employment_occupation', 
                   'health_insurance', 'census_msa', 'rent_or_own', 
                   'household_adults', 'household_children']
df.drop(columns=columns_to_drop, inplace=True)
df.columns
```

2. **Check missing values**

```python
# lets identify the missing values
missing_values = df.isna().sum()
missing_values
```

```python
# let's check the percentage of missing values
# Calculate the percentage of missing values
missing_percentage = (df.isna().sum() / len(df)) * 100

# Create a DataFrame to display both count and percentage of missing values
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

# Print the missing data DataFrame
print(missing_data)
```

The 'income_poverty' has a high % of missing data at 16%, followed by 'doctor_recc_h1n1' and 'doctor_recc_seasonal' where both have 8% of the same. While income might influence vaccination behavior, the large amount of missing data makes it less practical to include.
```python
# Lets drop the three columns with high missing values.
col_drop = ['income_poverty', 'doctor_recc_h1n1', 'doctor_recc_seasonal']
df.drop(columns=col_drop, inplace=True)
```

```python
# Check if the missing columns have been dropped
df.columns
```

```python
# Lets fill in the missing values using fillna
# Identify numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values in numerical columns with the mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill missing values in categorical columns with 'unknown'
df[categorical_cols] = df[categorical_cols].fillna('unknown')
df
```

Filled in the missing numerical values with the mean which ensures that they are filled with a neutral value. The categorical columns i used unknown so that it is not biased in any way.

```python
# Check if there are any missing values
df.isna().sum()
```

```python
# Check for duplicates

duplicates = df.duplicated()

# Count the number of duplicates
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")

# Display duplicate rows
duplicate_rows = df[duplicates]

print("Duplicate rows:")
print(duplicate_rows)
```

I checked and identified rows with identical values across all columns. Removed the duplicates from the dataset and confirmed the shape of the dataset post removal of duplicates. Removing duplicates ensures that each observation is counted only once.

4. **Unique Values**

```python
df.nunique()
```

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f'{column}: {unique_values}')
```

```python
# Handling the unique values
# Define columns that should contain only integers
integer_columns = [
    'h1n1_concern', 'h1n1_knowledge', 'behavioral_antiviral_meds', 'behavioral_avoidance',
    'behavioral_face_mask', 'behavioral_wash_hands', 'behavioral_large_gatherings',
    'behavioral_outside_home', 'behavioral_touch_face', 'chronic_med_condition',
    'child_under_6_months', 'health_worker', 'opinion_h1n1_vacc_effective',
    'opinion_h1n1_risk', 'opinion_h1n1_sick_from_vacc', 'opinion_seas_vacc_effective',
    'opinion_seas_risk', 'opinion_seas_sick_from_vacc'
]

# Round non-integer (float) values to the nearest whole number
df[integer_columns] = df[integer_columns].apply(lambda x: x.round())

# Verify the changes
for column in integer_columns:
    print(f'{column}: {df[column].unique()}')

# Handle 'unknown' values in categorical columns
categorical_columns_with_unknowns = ['education', 'marital_status', 'employment_status']

# Replace 'unknown' with the mode of the column
for column in categorical_columns_with_unknowns:
    mode_value = df[column].mode()[0]
    df[column] = df[column].replace('unknown', mode_value)

# Verify changes in categorical columns
for column in categorical_columns_with_unknowns:
    print(f'{column}: {df[column].unique()}')

# Verify all unique values
unique_values = df.nunique()
print(unique_values)
```

After imputing the missing values with mean and 'unknown' this brought about unique values. So the imputed mean I rounded it off to the nearest integer this is to maintain the integrity of originally integer-valued columns. Replace 'unknown' values with the mode of each categorical column to maintain consistency and avoid treating 'unknown' as a distinct category unless necessary. 

5. **Checking for Outliers**

To check for outliers we can use statistical methods, or visualization in the numerical columns of the dataset. 

```python
# Let's identify our numerical variables
num_cols
```

```python
# Use Boxplots for outlier detection
# Plot boxplots for numerical columns to visualize outliers
plt.figure(figsize=(15, 20))
for i, col in enumerate(num_cols, 1):
    plt.subplot(7, 3, i)
    sns.boxplot(df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Use IQR method to detect outliers
# Calculate IQR for numerical columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using IQR (1.5 times the IQR from Q1 and Q3)
outliers_iqr = df[((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print("\nOutliers detected using IQR:")
print(outliers_iqr)
```

The IQR had detected several outliers in the dataset. Lets' cap the outliers as will replace extreme values with a threshold value.

**Capping Outliers**

```python
# Calculate IQR for numerical columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Define the caps
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers
for col in num_cols:
    df[col] = np.where(df[col] < lower_bound[col], lower_bound[col], df[col])
    df[col] = np.where(df[col] > upper_bound[col], upper_bound[col], df[col])

# Verify the changes
print(df[num_cols].describe())
```

2. **Bivariate Analysis**
   Let's perform bivariate analysis using seasonal_vaccine as the target variable. We'll explore its relationship with other variables in the dataset.

```python
# Prepare data: Remove "seasonal_vaccine" from the dataframe
outcome_vars = df.drop(columns=['seasonal_vaccine']).columns

# Set up the plot
num_plots = 2
num_rows = (len(outcome_vars) + num_plots - 1) // num_plots
fig, axes = plt.subplots(num_rows, num_plots, figsize=(15, 5*num_rows))

# Create visualizations for each outcome variable
for i, var in enumerate(outcome_vars):
    row = i // num_plots
    col = i % num_plots
    ax = axes[row, col]
    
    # Plot stacked bar plot
    sns.countplot(data=df, x=var, hue='seasonal_vaccine', ax=ax)
    ax.set_title(f'Seasonal Vaccine Distribution by {var}')
    ax.set_xlabel(var)
    ax.set_ylabel('Count')
    ax.legend(title='Seasonal Vaccine', loc='upper right')

# Adjust layout
plt.tight_layout()
plt.show()
```

There is a co-relation between seasaonal vaccine and the other variables this can be seen in the analysis above.
